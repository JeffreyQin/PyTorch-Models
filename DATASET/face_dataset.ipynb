```python
class FaceLandmarksDataset(utils.data.Dataset):
    def __init__(self, csv_file, root_dir, transform=None):
        self.landmarks_df = pd.read_csv(csv_file)
        self.root_dir = root_dir
        self.transform = transform

    def __len__(self):
        return len(self.landmarks_df) # num of rows in DataFrame

    def __getitem__(self, idx):
        if torch.is_tensor(idx):
            idx = idx.tolist()

        img_path = os.path.join(self.root_dir, self.landmarks_df.iloc[idx, 0])
        image = io.imread(img_path)
        landmarks = self.landmarks_df.iloc[idx, 1:]
        landmarks = np.array([landmarks], dtype=float).reshape(-1,2)
        sample = {'image': image, 'landmarks': landmarks}

        if self.transform:
            sample = self.transform(sample)

        return sample
```

```python
face_dataset = FaceLandmarksDataset(csv_file='faces/face_landmarks.csv', root_dir='faces/')

for idx, sample in enumerate(face_dataset):
    print(idx, sample['image'].shape, sample['landmarks'].shape)

    ax = plt.subplot(1,4,idx+1)
    plt.tight_layout()
    ax.set_title('Sample {}'.format(idx))
    ax.axis('off')
    show_image(**sample)

    if idx == 3:
        plt.show()
        break
    
```

```python
# rescale images to the same size

class Rescale(object):
    def __init__(self, output_size):
        assert isinstance(output_size, (int, tuple))
        self.out_size = output_size

    def __call__(self, sample):
        image, landmarks = sample['image'], sample['landmarks']
        h, w = image.shape[:2]
        if isinstance(self.out_size, int):
            if h > w:
                new_h, new_w = self.output_size * h / w, self.output_size
            else:
                new_h, new_w = self.output_size, self.output_size * w / h
        else:
            new_h, new_w = self.output_size

        new_h, new_w = int(new_h), int(new_w)

        img = transform.resize(image, (new_h, new_w))

        return {'image': image, 'landmarks': landmarks}
```

```python
# convert numpy image to tensor

class ToTensor(object):
    def __call__(self, sample):
        image, landmarks = sample['image'], sample['landmarks']

        image = image.transpose((2,0,1))
        return {"image": torch.from_numpy(image), 'landmarks': torch.from_numpy(landmarks)}
```
